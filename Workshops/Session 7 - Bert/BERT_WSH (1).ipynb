Remove extra punctuation:

```python
import pandas as pd

# Define characters to remove as punctuation
persian_punctuation = "،/؛ـ«»." #"،?/؛ـ«»؟.!"

# Function to remove extra punctuation
def remove_extra_punctuation(text):
  # Convert text to string (in case it's not already)
  text = str(text)
  # Replace multiple occurrences of punctuation with a single space
  return text.replace("  ", " ").translate(str.maketrans('', '', persian_punctuation))

# Apply the function to the 'content' column
train['content_clean_punctuation'] = train['content'].apply(remove_extra_punctuation)
```

Convert Arabic letters to Persian ones and standardization:

```python
import pandas as pd

# Define a dictionary for character mapping
persian_char_map = {
  'ك': 'ک', 'ک': 'ک',
  'دِ': 'د', 'بِ': 'ب', 'زِ': 'ز', 'ذِ': 'ذ',
  'شِ': 'ش', 'سِ': 'س', 'ى': 'ی', 'ي': 'ی', 'ئ': 'ی',
  '١': '۱', '٢': '۲', '٣': '۳', '٤': '۴', '٥': '۵',
  '٦': '۶', '٧': '۷', '٨': '۸', '٩': '۹', '٠': '۰'
}

# Function to convert characters
def convert_arabic_to_persian(text):
  text = str(text)  # Ensure text is a string
  for arabic, persian in persian_char_map.items():
    text = text.replace(arabic, persian)
  return text

# Apply the function to the 'content' column
train['content_clean'] = train['content_clean_punctuation'].apply(convert_arabic_to_persian)
```

```python
train['word_count'] = train['content_clean'].apply(lambda x: len(str(x).split()))
```

```python
min_max_len = train['word_count'].min(), train['word_count'].max()
print(f'Min: {min_max_len[0]} \tMax: {min_max_len[1]}')
```

```python
def data_gl_than(data, less_than=100.0, greater_than=0.0, col='word_count'):
    data_length = data[col].values

    data_glt = sum([1 for length in data_length if greater_than < length <= less_than])

    data_glt_rate = (data_glt / len(data_length)) * 100

    print(f'Texts with word length of greater than {greater_than} and less than {less_than} includes {data_glt_rate:.2f}% of the whole!')

data_gl_than(train, 500, 100)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Histogram(
    x=train['word_count']
))

fig.update_layout(
    title_text='Distribution of word counts within comments',
    xaxis_title_text='Word Count',
    yaxis_title_text='Frequency',
    bargap=0.2,
    bargroupgap=0.2)

fig.show()
```
