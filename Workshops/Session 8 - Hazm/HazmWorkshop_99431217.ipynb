# Counting the number of verbs and adverbs

**Harf:**

```python
from hazm import POSTagger
tagger = POSTagger(model='pos_tagger.model')
Harf_Tags = tagger.tag(word_tokenize(Harf_Output_Normalized))
```

```python
Harf_Tags
```

```python
Harf_verb_count = 0
Harf_adverb_count = 0
for token, tag in Harf_Tags:
        if tag.startswith('VERB'):  # Verbs start with 'VB'
            Harf_verb_count += 1
        elif tag.startswith('ADV'):  # Adverbs start with 'RB'
            Harf_adverb_count += 1
```

```python
print("Number of verbs in Harf: ", Harf_verb_count)
print("Number of adverbs in Harf: ", Harf_adverb_count)
```

```python
Original_verb_count = 0
Original_adverb_count = 0
for token, tag in Original_Tags:
        if tag.startswith('VERB'):  # Verbs start with 'VB'
            Original_verb_count += 1
        elif tag.startswith('ADV'):  # Adverbs start with 'RB'
            Original_adverb_count += 1
```

```python
print("Number of verbs in Original: ", Original_verb_count)
print("Number of adverbs in Original: ", Original_adverb_count)
```

# Root of the verb with the most appearance

**Harf:**

```python
from collections import Counter
def find_most_frequent_verb(text_with_tags):
    # Tokenization
    # tokens_with_tags = [tuple(word_tag.split(',')) for word_tag in text_with_tags]
    tokens_with_tags = text_with_tags
    # Filtering and Counting Verbs
    verbs = [token for token, tag in tokens_with_tags if tag.startswith('VERB')]
    verb_counts = Counter(verbs)

    # Finding the most frequent verb
    most_frequent_verb = verb_counts.most_common(1)[0][0]

    return most_frequent_verb
```

```python
most_frequent_verb_harf = find_most_frequent_verb(Harf_Tags)
print("Most frequent verb in Harf:", most_frequent_verb_harf)
```

```python
from hazm import Lemmatizer
lemmatizer_harf = Lemmatizer()
lemmatizer_harf.lemmatize (most_frequent_verb_harf)
```

```python
most_frequent_verb_original = find_most_frequent_verb(Original_Tags)
print("Most frequent verb in Original:", most_frequent_verb_original)
```

```python
from hazm import Lemmatizer
lemmatizer_harf = Lemmatizer()
lemmatizer_harf.lemmatize (most_frequent_verb_original)
```
